```python
#### 75. How to compute averages using a sliding window over an array? (★★★)
#[1,2,3,4,5,6,7,8]  n = 3
def slide(a, n):
    cum_sum = np.cumsum(a,dtype=np.float32)
    cum_sum[n:] = cum_sum[n:] - cum_sum[:-n]
    cum_sum = cum_sum[n-1:] / n
    return cum_sum
slide(np.arange(20), 3)
```

```python
#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?
x = np.array([0,3,3])
y = np.array([0,0,1])

dist = np.cumsum(np.sqrt(np.diff(x)**2 + np.diff(y)**2))
dist = np.insert(dist, 0, 0)
equidistant = np.linspace(0,dist[-1],len(x)+1)
new_x = np.interp(equidistant, dist,x) 
new_y = np.interp(equidistant, dist, y)
new_x, new_y
```
